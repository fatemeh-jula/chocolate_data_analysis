خواندن فایل به صورت دیتافریم

```python
import numpy as np
import pandas as pd

df = pd.read_csv('chocolate.csv')
df.head()
```

بررسی ابعاد دیتافریم

```python
s = np.shape(df)
cols = df.columns
print(s)
print(cols)
```

مرتب کردن اسم ستون ها

```python
df.rename(columns={"Specific Bean\nOrigin":"Specific Bean Origin","Review\nDate":"Review Date","Cocoa\nPercent":"Cocoa Percent",
                                "Company\nLocation":"Company Location","Bean\nType":"Bean Type","Broad Bean\nOrigin":"Broad Bean Origin"}, inplace=True)
df.head()
```

نگاه کلی به دیتافریم

```python
df.info()
```

تغییر مقدار ستون ها به رشته متنی

```python
df[["Company","Specific Bean Origin","Cocoa Percent","Company Location","Bean Type","Broad Bean Origin"]] = df[["Company","Specific Bean Origin","Cocoa Percent","Company Location","Bean Type","Broad Bean Origin"]].astype('string')
df.info()
```

```python
df['Cocoa Percent'] = df['Cocoa Percent'].str.replace("%","").astype(float)
df['Cocoa Percent']
```

به درصد rating تبدیل مقادیر ستون

```python
df['Rating'] = df['Rating'] * 20

df.head()
```

محاسبه قیمت 100 گرم از هر نوع شکلات

```python
df['price(100g)'] = 25 * df["Rating"] * df["Cocoa Percent"]
df
```

جدا کردن شکلات های تلخ

```python
dark_chocolates = df[df["Cocoa Percent"]>70].reset_index(drop=True)
dark_chocolates
```

بررسی وجود دانه ای خاص در شکلات ها

```python
has_Trinitario = dark_chocolates["Bean Type"].str.contains("Trinitario")
has_Trinitario
```

افزایش 10 درصدی شکلات های دارای دانه خاص

```python
dark_chocolates.loc[has_Trinitario, 'price(100g)'] = dark_chocolates.loc[has_Trinitario, 'price(100g)'] * 1.1
dark_chocolates
```

مبلغ دریافتی از فروش این شکلات ها

```python
priceSum = dark_chocolates["price(100g)"].sum()
priceSum
```

جداسازی شکلات های غیر تلخ

```python
df = df[df["Cocoa Percent"]<=70].reset_index(drop=True)
df
```

به ازای هر شرکت، میانگین امتیاز شکلات های آن برای هرسال

```python
companies = df.groupby(["Company","Review Date"])["Rating"].mean().unstack()
companies.head()
```

```python
companies.columns.name = None
companies.index.name = None
```

پیدا کردن بهترین شرکت ها، از سال 2012 به بعد

```python
mean_ratings =companies[[2012,2013,2014,2015,2016,2017]].mean(axis=1)
mean_ratings
```

```python
a = mean_ratings.copy()
df_sorted = a.reset_index(name="ave")
df_sorted = df_sorted.sort_values(by=["ave","index"],ascending=[False,True])
best_ratings =df_sorted.set_index("index").head(10)["ave"]
best_ratings
```

پیدا کردن شکلات های 10 شرکت برتر

```python
chocolates_to_sell = df[df["Company"].isin(best_ratings.index)]
chocolates_to_sell = chocolates_to_sell.reset_index(drop=True)
chocolates_to_sell
```
